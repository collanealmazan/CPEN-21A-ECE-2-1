##For Loop

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
```

##The Break Statement


```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  print(x)
  if x=="Thursday":
    break
```

```python
week = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
for x in week:
  if x=="Thursday":
    break
  print(x)
```

##Looping Through String


```python
for x in "Python Programming":
  print(x)
```

##The Range Function

```python
for x in range(16):
  print(x)
```

```python
for x in range(2,16):
  print(x)
```

##Nested Loops

```python
adjective = ["red","big","tasty"]
fruits = ["apple","banana","cherry"]
for x in adjective:
  for y in fruits:
    print(x,y)
```

##While Loop

```python
i = 1
while i<=6:
  print(i)
  i+=1 #Assignment operator for addition
```

# The break statement

```python
i=1
while 1<6:
  print(i)
  if i==3:
    break
  i+=1
```

# The continue statement

```python
i = 0
while i<6:
  i+=1 #Assignment Operator for addition
  if i==3:
    continue
  print(i)
```

#The else statement

```python
i = 1
while i<=6:
  print(i)
  i+=1
else:
  print("i is no longer less than 6")
```
