```python
a = 30
b = 30
if a>b:
  print("a is greater than b")
elif b>a:
  print("b is greater than a")
else:
  print("a is equal to b")
```

##Nested if... Else Statement

```python
x = 20

if x>10:
  print("Above ten")
  if x>20:
    print("Above twenty")
  else:
    print("Above ten but Not above twenty")
else: 
  print("Not above ten")
```
