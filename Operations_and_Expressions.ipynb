##Boolean Operator


```python
x=1
y=2

print(x>y)
print(10>11)
print(10==10)
print(10!=11)
```

```python
#using bool() function

print(bool("Hello"))
print(bool(15))

print(bool(True))
print(bool(False))
print(bool(None))
print(bool(0))
print(bool([]))
```

###Functions can return Boolean


```python
def myFunction(): return True

print(myFunction())
```

```python
def yourFunction(): return False

if yourFunction():
  print("Yes!")
else: 
  print("No!")
```

##Arithmetic Operators


```python
print(10+5)
print(10-5)
print(10*5)
print(10/5)
print(10%5) #modulo division, remainder
print(10//5) #floor division
print(10//3) #floor division
print(10%3) #3 x 3 = 9
```

##Bitwise Operators


```python
a = 60 # 0011 1100
b = 13 # 0000 1101

print(a&b)
print(a|b)
print(a^b)
print(~a)
print(a<<1) # 0111 1000
print(a<<2) # 1111 0000
print(b>>1) # 1 0000 0110
print(b>>2) # 0000 0011  carry flag bit = 01  
```

##Logical Operators

```python
#and logical operator

a = True
b = False

print(a and b)
print(not(a and b))
print(a or b)
print(not(a or b))
```
